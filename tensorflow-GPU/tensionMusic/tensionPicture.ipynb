```python
fig = plt.figure()
ax = fig.gca(projection="3d")
t = np.linspace(-3*np.pi,3*np.pi,100)
z = t
x = np.sin(t)
y = np.cos(t)
ax.plot(x,y,z,label='spiral array')
ax.legend()
plt.show()
```
